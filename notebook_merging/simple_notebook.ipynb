```python
import numpy as np
import pandas as pd

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
x = np.linspace(0,4)
y = np.sin(x)
```

```python
p1 = bokeh.plotting.figure(width = 450, height = 350)
p1.line(x,y, line_width = 2)
bokeh.io.show(p1)
```

```python
p2 = bokeh.plotting.figure(width = 450, height = 350)
p2.line(y,x, line_width = 2)
bokeh.io.show(p2)
```
